### Defino funcion para descartar los elementos que no sean iguales a Y o N y asignarles el valor de desconocidos.

```python
def limpiarfatal(x):
    for e in x :
        if e == 'Y' or e == 'N':
            return e
        else:
            return 'UNKNOWN'

    
```

```python
tibu["Fatal (Y/N)"].apply(limpiarfatal).unique() #No me deja iterar con valores float, voy a cambiarlos todos a string
```

```python
tibu["Fatal (Y/N)"].apply(type).value_counts()
```

```python
tibu["Fatal (Y/N)"] = tibu["Fatal (Y/N)"].astype(str)   #Cambio valores de la columna a string
```

```python
tibu["Fatal (Y/N)"].apply(limpiarfatal).unique() # Done it, ya solo queda quitar los nulos, en este caso los reemplazare por desconocido.
```

```python
tibu["Fatal (Y/N)"]= tibu["Fatal (Y/N)"].apply(limpiarfatal)
```
